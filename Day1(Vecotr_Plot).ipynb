```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define the vectors
features = ['Sky', 'Size', 'Color']
aeroplane_vector = np.array([1, 5, 3])
bird_vector = np.array([1, 2, 4])
```

```python
# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
# Plot the vectors
ax.quiver(0, 0, 0, aeroplane_vector[0], aeroplane_vector[1], aeroplane_vector[2], color='r', label='Aeroplane', arrow_length_ratio=0.1)
ax.quiver(0, 0, 0, bird_vector[0], bird_vector[1], bird_vector[2], color='b', label='Bird', arrow_length_ratio=0.1)
# Annotate the points
ax.text(aeroplane_vector[0], aeroplane_vector[1], aeroplane_vector[2], "Aeroplane", color='r')
ax.text(bird_vector[0], bird_vector[1], bird_vector[2], "Bird", color='b')
# Set labels and title
ax.set_xlabel('Sky')
ax.set_ylabel('Size')
ax.set_zlabel('Color')
ax.set_title('3D Representation of Document Vectors')


# Set limits for better visibility
ax.set_xlim(0, 6)
ax.set_ylim(0, 6)
ax.set_zlim(0, 6)

# Show legend and plot
ax.legend()
plt.grid()
plt.show()
```
